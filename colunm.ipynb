```python
# Calculate cross-sectional area and moment of inertia for square and round tubes
import numpy as np

tubes_df = tubes_df.copy()
w = tubes_df["Width (m)"]  # For round tubes, this is diameter
t = tubes_df["Wall Thickness (m)"]

# Initialize arrays for area and moment of inertia
area = np.zeros(len(tubes_df))
moment_of_inertia = np.zeros(len(tubes_df))

for i, tube_type in enumerate(tubes_df["Tube Type"]):
    if tube_type == "Square tube":
        # Square tube formulas
        # Area = (width^2 - (width - 2*thickness)^2)
        # Moment of inertia (I) = (width^4 - (width - 2*thickness)^4) / 12
        width = w.iloc[i]
        thickness = t.iloc[i]
        area[i] = width**2 - (width - 2*thickness)**2
        moment_of_inertia[i] = (width**4 - (width - 2*thickness)**4) / 12
    
    elif tube_type == "Round tube":
        # Round tube formulas (hollow circular cross-section)
        # Area = π * (R_outer^2 - R_inner^2) = π/4 * (D_outer^2 - D_inner^2)
        # Moment of inertia (I) = π/64 * (D_outer^4 - D_inner^4)
        diameter_outer = w.iloc[i]
        thickness = t.iloc[i]
        diameter_inner = diameter_outer - 2*thickness
        area[i] = np.pi/4 * (diameter_outer**2 - diameter_inner**2)
        moment_of_inertia[i] = np.pi/64 * (diameter_outer**4 - diameter_inner**4)

tubes_df["Cross Sectional Area (m²)"] = area
tubes_df["Moment of Inertia (m⁴)"] = moment_of_inertia

# Merge alloy details
tubes_mechanical_df = tubes_df.merge(alloys_df, left_on="Alloy", right_on="Alloy", how="left")
tubes_mechanical_df
```

```python
import numpy as np

# Calculate buckling load using Euler's formula
# P_critical = (π² × E × I) / (K × L)²
# Where K is the end condition factor and L is the column length

buckling_df = tubes_mechanical_df.copy()

# Extract values for calculation
E = buckling_df["Modulus of Elasticity (Pa)"]
I = buckling_df["Moment of Inertia (m⁴)"]
K = END_CONDITION  # End condition factor from global constants
L = COLUMN_HEIGHT  # Column length from global constants

# Calculate critical buckling load (Euler load)
buckling_df["Critical Buckling Load (N)"] = (np.pi**2 * E * I) / (K * L)**2

# Calculate allowable load with factor of safety
buckling_df["Allowable Load (N)"] = buckling_df["Critical Buckling Load (N)"] / FACTOR_OF_SAFETY

# Display results
buckling_df[["Tube Type", "Alloy", "Critical Buckling Load (N)", "Allowable Load (N)"]]
```

```python
mass_cost_df = buckling_df.copy()

# Calculate mass = density × volume = density × cross_sectional_area × length
mass_cost_df["Volume (m³)"] = mass_cost_df["Cross Sectional Area (m²)"] * COLUMN_HEIGHT
mass_cost_df["Mass (kg)"] = mass_cost_df["Density (kg/m³)"] * mass_cost_df["Volume (m³)"]

# Calculate total cost: use Actual Cost if available, otherwise calculate from mass × cost_per_kg
mass_cost_df["Calculated Cost (USD)"] = mass_cost_df["Mass (kg)"] * mass_cost_df["Approx Cost (USD/kg)"]
mass_cost_df["Single Column Cost (USD)"] = mass_cost_df["Actual Cost"].fillna(mass_cost_df["Calculated Cost (USD)"])
mass_cost_df["Total Cost (USD)"] = mass_cost_df["Single Column Cost (USD)"] * COLUMN_COUNT

# Display mass and cost results
mass_cost_df[["Tube Type", "Alloy", "Mass (kg)", "Single Column Cost (USD)", "Total Cost (USD)"]]
```

```python
# Create packing table - assumes columns are packed in a 2x2 square arrangement
packing_df = mass_cost_df.copy()

# Calculate total mass for all columns
packing_df["Total Packed Mass (kg)"] = packing_df["Mass (kg)"] * COLUMN_COUNT

# Calculate packing dimensions (2x2 arrangement)
# For both square and round tubes, use the width/diameter for container sizing
# Width and height of the square container = 2 * tube width/diameter
packing_df["Container Width (m)"] = 2 * packing_df["Width (m)"]
packing_df["Container Height (m)"] = 2 * packing_df["Width (m)"]  # Square container
packing_df["Container Length (m)"] = COLUMN_HEIGHT  # Same as column height

# Display packing results
packing_columns = [
    "Tube Type", "Alloy", "Total Packed Mass (kg)", 
    "Container Width (m)", "Container Height (m)", "Container Length (m)"
]
packing_df[packing_columns]
```

```python
# Create a summary table with selected columns
summary_columns = [
    "Tube Type", "Alloy", "Width (m)", "Wall Thickness (m)",
    "Mass (kg)", "Single Column Cost (USD)", "Total Cost (USD)", 
    "Allowable Load (N)"
]

mass_cost_summary = mass_cost_df[summary_columns].copy()

# Add packed mass and container width from packing_df
mass_cost_summary["Total Packed Mass (kg)"] = packing_df["Total Packed Mass (kg)"]
mass_cost_summary["Container Width (m)"] = packing_df["Container Width (m)"]

mass_cost_summary
```

```python
# Create Imperial summary table with converted units
imperial_summary = mass_cost_summary.copy()

# Conversion factors
m_to_inch = 1 / inch_to_m  # 39.3701 inches per meter
kg_to_lb = 2.20462  # pounds per kilogram
N_to_lbf = 0.224809  # pound-force per Newton

# Convert metric to imperial units
imperial_summary["Width (in)"] = imperial_summary["Width (m)"] * m_to_inch
imperial_summary["Wall Thickness (in)"] = imperial_summary["Wall Thickness (m)"] * m_to_inch
imperial_summary["Mass (lb)"] = imperial_summary["Mass (kg)"] * kg_to_lb
imperial_summary["Allowable Load (lbf)"] = imperial_summary["Allowable Load (N)"] * N_to_lbf
imperial_summary["Total Packed Mass (lb)"] = imperial_summary["Total Packed Mass (kg)"] * kg_to_lb
imperial_summary["Container Width (in)"] = imperial_summary["Container Width (m)"] * m_to_inch

# Select imperial columns for display
imperial_columns = [
    "Tube Type", "Alloy", "Width (in)", "Wall Thickness (in)",
    "Mass (lb)", "Single Column Cost (USD)", "Total Cost (USD)", 
    "Allowable Load (lbf)", "Total Packed Mass (lb)", "Container Width (in)"
]

imperial_summary_display = imperial_summary[imperial_columns]
imperial_summary_display
```

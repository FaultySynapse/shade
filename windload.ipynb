Wind speed estimates come from section 1609.1.1 of the 2018 Northern Nevada Code Amendments.  
https://nnicc.org/wp-content/uploads/2025/03/2024-NNICC-Amendments-Final-3-12-25.pdf

| Risk Category | Ultimate Wind Speed V<sub>ult</sub> (mph) | Vasd Wind Speed 3-sec gust (mph) |
|---------------|------------------------------------------|-----------------------------------|
| I             | 110                                      | 85                                |
| II & 2024 IRC | 120                                      | 93                                |
| III           | 130                                      | 101                               |
| IV            | 135                                      | 104                               |

This is higher than the 97 mph called for by the ASCE 7 wind speed map.  
Risk category is 2 for normal. Exposure Category C for flat open space. No adjustments.

```python
WIND_SPEED = 40.23 # m/s (90 mph)
WALL_HEIGHT = 2.5 # m (8.2 ft)
ROOF_HEIGHT = 1 # m (3.28 ft)
TOTAL_HEIGHT = WALL_HEIGHT + ROOF_HEIGHT # m
```

```python
width_options = [5, 10] # m
Cd = [0.9, 1.2, 1.7]  # Cd for ideal 45 wedge, taut flat surface, and for flexible vertical one
```

We are estimating the wind load on the structure using the simplified formula:

$$
F = 0.5 \cdot \rho \cdot V^2 \cdot C_d \cdot A
$$

where:

- $F$ = wind load (N)
- $\rho$ = air density (kg/m³), assumed as 1.225 kg/m³
- $V$ = wind speed (m/s), set to 50 m/s
- $C_d$ = drag coefficient, varies based on surface type (values: 0.9, 1.2, 1.7)
- $A$ = projected area facing the wind (m²), calculated as width × height

We consider different structure widths (`width_options`) and drag coefficients (`Cd`) to compute a range of possible wind loads. The results are summarized in the `wind_load_df` table, showing the wind load for each combination of width and drag coefficient. This approach provides a quick estimate for design and comparison purposes.

```python
import pandas as pd

# Calculate approximate wind load: F = 0.5 * rho * V^2 * Cd * A
rho = 1.225  # air density in kg/m^3
wind_loads = []
for width in width_options:
    area = width * TOTAL_HEIGHT 
    for cd in Cd:
        F = 0.5 * rho * WIND_SPEED**2 * cd * area
        wind_loads.append({'width': width, 'Cd': cd, 'wind_load_N': F})

wind_load_df = pd.DataFrame(wind_loads)
wind_load_df
```
